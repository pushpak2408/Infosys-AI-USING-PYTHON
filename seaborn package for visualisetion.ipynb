```python
mca=sns.load_dataset('titanic')
mca.head()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='sex',data=mca)
```

```python
mca.shape
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='sex', data=mca, hue='survived', palette='Set2')
```

```python
#kde plot
plt.figure(figsize=(3,3))
sns.kdeplot(x='age',data=mca,color='black')
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(x='pclass',data=mca,color='black')
```

```python
#distribution plot
plt.figure(figsize=(3,3))
sns.displot(x='age', kde=True,bins=5,data=mca)
```

```python
sns.displot(x='age', kde=True, bins=5, data=mca, hue='survived', palette='Set3')
```

```python
# scatter plots
mca= sns.load_dataset('iris')
mca.head()
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x='sepal_length', y='petal_length', data=mca, hue='species')
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='sepal_width', y='petal_width', data=mca, hue='species')
```

```python
mca.shape
```

```python
#joint plot
plt.figure(figsize=(3,3))
sns.jointplot(x='sepal_length',y='petal_length', data=mca,kind='reg',color='green')
```

```python
#joint plot
plt.figure(figsize=(3,3))
sns.jointplot(x='sepal_length',y='petal_length', data=mca,hue='species',color='green')
```

```python
#pairplot

sns.pairplot(mca)
```

```python
import pandas as pd
# Assuming `mca` is your DataFrame
numeric_df = mca.select_dtypes(include=[float, int])
# Calculate the correlation matrix
corr = numeric_df.corr()
# Display the correlation matrix
print(corr)
```

```python
#heatmap
plt.figure(figsize=(5,3))
sns.heatmap(corr,annot=True)
```

```python
plt.figure(figsize=(8,3))
sns.heatmap(corr, annot=True,cmap=['red','green','blue'], vmin=-.5,vmax=0.6)
```
