```python
import matplotlib.pyplot as plt

# Sample data
empid = np.arange(1, 10)
salary = np.arange(20, 110, 10)
plt.figure(figsize=(3,3))
ax = plt.axes()
ax.set_facecolor('#ECF0F1')
plt.bar(empid, salary, color='#FFA726')
plt.xlabel("empid")
plt.ylabel("salary")
plt.show()
```

```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,29,40]
plt.figure(figsize=(3,3))
ax = plt.axes()
ax.set_facecolor('white')
plt.bar(x1, y1, label='first', color='green')
plt.bar(x2, y2, label='second', color='blue')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$Bar$ $Chart$')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
barplot= plt.bar(empid,salary)
for bar in barplot:
    yval=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0,yval,int(yval), va= 'bottom')
plt.title('Bar Graph Example')
plt.xlabel('emp')
plt.ylabel('sal')
plt.show()
```

## Stack Bar charts

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

y1 = np.array([1, 3, 5, 7, 9])
y2 = np.array([2, 4, 6, 8, 10])
y3 = np.array([20, 40, 60, 80, 100])

barwidth = 0.25
pos1 = np.arange(len(y1))
pos2 = [x + barwidth for x in pos1]
pos3 = [x + barwidth for x in pos2]

plt.bar(pos1, y1, color='blue', width=barwidth, label='Batting')
plt.bar(pos2, y2, color='green', width=barwidth, label='Bowling')
plt.bar(pos3, y3, color='red', width=barwidth, label='Fielding')

plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of tickets', fontweight='bold')
plt.xticks([i + barwidth for i in range(len(y1))], ['Sachin', 'Saurav', 'Rahul', 'Dhoni', 'Virat'])
plt.legend()
plt.show()

```

### multiple line graph

```python
import pandas as pd
mca= pd.DataFrame({'year':[2016,2017,2018, 2019,2022],'sales':[2000,3000,4000,5000,6000], 'revenue':[1000,1500,2000,2500,3000]})
mca
```

```python
# multiline plot
plt.figure(figsize=(2,2))
ax=mca.plot('year','sales', kind= 'line', label='sales per year')
mca.plot('year','revenue', kind= 'line', label='revenue per year')
ax.set(ylabel='sales', xlabel='year', xticks=mca['year'])
```
